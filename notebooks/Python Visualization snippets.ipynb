### If we have three experiments that take time: `time`, `time ** 2` and `exp(time)` seconds

```python
time = np.arange(0., 5., 0.2)
```

```python
time
```

```python
time2 = time ** 2
print(time2)
```

```python
time3 = np.exp(time)
print(time3)
```

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.show()
```

### Nicer bigger and clearer plot

### But there are lot of things missing, we'll try to answer a few:

1. Looking just at the plot without the code, I don't know what the plot is about?
2. What does the x-axis and y-axis mean?
3. What color represents what?

### 1. Looking just at the plot without the code, I don't know what the plot is about?

To answer it, `matplotlib` provides way to put titles to a plot, how? let's find out:

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.show()
```

### 2. What does the x-axis and y-axis mean?

To answer it, `matplotlib` provides `xlabels` and `ylabels`

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.show()
```

### Q: Can you try out, `ylabels`?

### 3. What color represents what?

We can try to answer this using `legends`

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
plt.show()
```

### There can be more cleanups to the above plot

* Can remove the ticks on `x-axes` (we don't need number of steps here and also `y-axes` too can be removed)

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
plt.xticks([])
plt.show()
```

* We can add grids at the background

```python
plt.plot(time)
plt.plot(time2)
plt.plot(time3)
plt.title('Time taken for Python Processes', fontdict={'fontsize': 15, 'fontweight': 10}, pad=15)
plt.xlabel('Number of steps', fontdict={'fontsize': 15}, labelpad=30)
plt.legend(['Regular', 'Squared', 'Exponential'], loc='best')
# plt.xticks([])
plt.grid(axis='both')
plt.show()
```
